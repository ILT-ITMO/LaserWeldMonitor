```python
import pandas as pd
import numpy as np
import glob
from sdv.single_table import GaussianCopulaSynthesizer
from sdv.metadata import Metadata
from sdv.evaluation.single_table import run_diagnostic, evaluate_quality
from sdv.evaluation.single_table import get_column_plot
import multiprocessing
```

```python
file_list = glob.glob('/content/*.xlsx')
```

```python
defect='hu'
```

```python
data=pd.read_excel(file_list[0], engine='openpyxl', sheet_name='spectras')
data1 = data.drop(columns=['hu','hg', 'he', 'hp', 'hs', 'hm', 'hi'], axis=1)  # Features
data1[defect]=np.where(data[defect] > 0, 1, 0)

data1.columns = data1.columns.astype(str)
```

```python
data1.head()
```

```python
len(data1)
```

```python
data1=data1.dropna()
```

```python
data1['hu'].value_counts()
```

```python
data_class=data1[data1['hu']==0]
```

```python
data_class['hu'].value_counts()
```

```python
metadata = Metadata.detect_from_dataframe(
    data=data_class,
)
```

```python
synthesizer = GaussianCopulaSynthesizer(metadata, default_distribution='norm')
synthesizer.fit(data_class)
num_samples = 100
synthetic_data_default = synthesizer.sample(num_rows=num_samples)
synthetic_data_default.to_excel("synthetic_data_class.xlsx", index=False)
```
